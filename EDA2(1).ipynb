 # Discuss the pros and cons of One-Hot Encoding and Label Encoding.

# One-Hot Encoding
One-Hot Encoding is a technique used to convert categorical variables into a format that can be provided to machine learning algorithms. It creates binary (0 or 1) columns for each category.
# Pros:
No Ordinal Relationship Assumed: Each category is treated as independent, with no order.

Suitable for Nominal Data: Works well for data like colors, gender, city names, etc.

Safe for Most Models: Prevents models from assuming a mathematical relationship between categories.
# Cons:
Increases Dimensionality: If a column has many unique values, it creates many new columns.

Consumes More Memory and Time: Because of the large number of columns.

May Lead to Sparse Data: Many zero values in the dataset, which can slow down training.






#  Label Encoding
Label Encoding converts each category into a unique number (e.g., Red = 0, Green = 1, Blue = 2). It’s simple and uses just one column.

# Pros:
Memory Efficient: Only one column is created instead of many.

Compact Representation: Keeps the dataset small and clean.
# Cons:
Assumes Ordinal Relationship: Models may think higher numbers mean higher importance, which is not always true.

Can Mislead Linear Models: Like Linear Regression, which interprets numerical differences as meaningful.



Works Well with Tree-Based Models: Such as Decision Tree, Random Forest, XGBoost.

# Use the Isolation Forest algorithm to identify and remove outliers. Discuss how outliers can affect model performance.

Isolation Forest is a machine learning algorithm that detects anomalies by randomly splitting data to isolate observations. The fewer splits needed to isolate a point, the more likely it is an outlier.



```python
from sklearn.ensemble import IsolationForest
```

```python
# training the model
df = df.select_dtypes(include=['number'])# Keeps only numeric columns
df_strings = df.select_dtypes(include=['object'])
```

```python
# Apply the Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)
model
```

```python
# Get predictions (these are -1 for anomalies and 1 for normal points)
outliers = model.fit_predict(df)
```

```python
outliers
```

```python
# The model returns 1 for normal points and -1 for outliers
df['outlier'] = outliers
df['outlier']
```

```python
# Remove the outliers
df_no_outliers = df[df['outlier'] == 1].drop('outlier', axis=1)  # Remove rows flagged as outliers
df_no_outliers 
```

```python
#Show the cleaned DataFrame
print(df_no_outliers.head())
```
